# Data Visualization using Matplotlib

### Importing Matplotlib PyPlot

```python
from matplotlib import pyplot as plt
```

### Single plot from points

```python
plt.plot([1,3,5,7],[1,9,25,49]) #Plots Y vs X by taking points list, first argument=X and second argument=Y
plt.show()
```

### Multiple plots on single graph

```python
x1 = [2,4,8]
y1 = [4,8,24]
```

```python
x2 = [3,6,9]
y2 = [10,15,30]
```

```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.title('Multiple Plots') #Gives title to the graph
plt.xlabel('X Axis') #Gives label to X Axis
plt.ylabel('Y Axis') #Gives label to Y Axis
plt.show()
```

### Bar Graph Visualization

```python
x1 = [2,4,6]
y1 = [3,6,9]
```

```python
x2 = [4,8,12]
y2 = [5,10,15]
```

```python
x3 = [6,12,18]
y3 = [7,14,21]
```

```python
plt.bar(x1,y1)
plt.bar(x2,y2)
plt.bar(x3,y3)
plt.title('Bar Graph')
plt.show()
```

### Histogram Visualization

```python
numbers=[4,12,36,42,23,41,15,17,19,64,86,24,55,84,97,92]
jumps=[0,10,20,30,40,50]
```

```python
plt.title('Histogram')
plt.hist(numbers,jumps,histtype='step')
plt.show()
```

### Scatter Plot Visualization

```python
x = [4,6,7,1,5,9]
y = [2,4,8,3,1,6]
```

```python
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.scatter(x,y)
```
