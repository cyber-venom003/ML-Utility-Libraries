# Data Visualization using Seaborn

### Importing Seaborn with other modules

```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

### Loading Database

```python
database = sb.load_dataset('diamonds') #Lists of datasets: https://github.com/mwaskom/seaborn-data
```

```python
print(database)
```

### Displot

```python
sb.displot(database['carat'],kde=True)
plt.show()
```

### Joint Plot

```python
tips_db = sb.load_dataset('tips')
print(tips_db)
```

```python
sb.jointplot(x='total_bill',y='tip',data=tips_db)
plt.show()
```

### Catplot

```python
flights_db = sb.load_dataset('flights')
print(flights_db)
```

```python
sb.catplot(x='month',y='passengers',data=flights_db,kind='strip') #Strip kind graph
plt.show()
```

```python
sb.catplot(x='month',y='passengers',data=flights_db,kind='violin') #Violin kind graph
plt.show()
```

```python
sb.catplot(x='month',y='passengers',data=flights_db,kind='box') #Box kind graph
plt.show()
```

### Multiple Graphs using FaceGrid

```python
graph = sb.FacetGrid(tips_db,col='smoker',hue='sex')
graph.map(plt.scatter,'total_bill','tip')
plt.show()
```
